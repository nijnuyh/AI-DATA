## Bonus Part
넷플릭스 데이터가 주어졌을 때, 다음 질문에 답하시오:

**가장 많은 작품이 올라간 국가는 어디이고, 얼마나 많은 작품이 있는가?**
- country column을 기준으로 한다.
- 단일 국가인 경우를 기준으로 결과를 구해보자.

먼저 문제 해결에 필요한 라이브러리를 호출한다.

```python
import numpy as np 
import pandas as pd 
from collections import Counter
```

주어진 링크를 통해 csv 파일을 다운받고 pd.read_csv 메소드를 활용하여 csv 파일에 접근한다.

```python
netflix_data = pd.read_csv("./netflix_titles.csv")
netflix_data
```

주어진 조건을 활용하여 country column의 값이 단일국가인 data 만을 이용한 DataFrame을 만든다.

```python
not_null = netflix_data[netflix_data['country'].isnull() == False]
single_country_data = not_null[~(not_null['country'].str.contains(","))] 

single_country_data
```

Counter 메소드를 이용하여 country column의 값을 기준으로 연산후, 가장 큰 값을 갖는 국가와 그 값을 출력

```python
count = Counter(single_country_data['country'])
print(count.most_common(1))
```

- 선의 색깔은 노란색(#f2a900) 으로 해야합니다.

💡 이동평균(Moving Average)법은 시계열 데이터를 표현하는 데에 많이 사용하는 방법 중 하나입니다.

먼저 문제 해결에 필요한 라이브러리를 호출한다.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

- 주어진 링크를 통해 `csv` 파일을 다운받고 `pd.read_csv` 메소드를 활용하여 `csv `파일에 접근한다.
- 주어진 날짜 조건에 맞게 DataFrame을 편집한다.
- Data를 날짜를 기준으로 정렬한 후 `index`를 재설정한다.

```python
bitcoin_data = pd.read_csv("./BitCoin.csv")
bitcoin_data = bitcoin_data[(bitcoin_data['Date']>='2016-06-01') & (bitcoin_data['Date']<'2017-07-01')]
bitcoin_data = bitcoin_data.sort_values('Date')
bitcoin_data.reset_index(inplace=True, drop=True)
bitcoin_data
```

`.to_datetime` 함수를 이용하여 `Date` column의 값을 형변환 해준다.

```python
bitcoin_data['Date'] = pd.to_datetime(bitcoin_data['Date'])
bitcoin_data
```

그래프를 그린다.

```python
ma5 = bitcoin_data['Open'].rolling(window=5).mean()
plt.rc('font', family='Malgun Gothic')
plt.figure(figsize=(15,8))
plt.plot(bitcoin_data['Date'], ma5, color="#f2a900")
plt.title(" MA(Moving Average) 비트코인 가격 그래프")
plt.xlabel("Date")
plt.ylabel("Bitcoin Moving Average")
plt.grid(True, linestyle='--')
plt.show()
```

해당 그래프를 본 나일론 마스크씨는 정신을 차렸다. 정확히는 비트코인만 바라본 본인을 말이다. 정말 많은 가상화폐가 있는데 비트코인만 바라본 본인을 돌아보게 되었다. 더불어 가장 핫한 코인인 이더리움 또한 관심이 생겼는데, 마스크씨는 2017년 한 해에 이 두개의 추세를 비교하고싶어한다. 마스크씨를 도와 비트코인과 이더리움의 가격 추세 비교를 도와주자.
